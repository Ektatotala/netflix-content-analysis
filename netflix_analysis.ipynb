```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("youssefkhaled117/netflix-titles-cleaned")

print("Path to dataset files:", path)
```

```python
import os
print(os.listdir(path))
```

```python
import pandas as pd

df = pd.read_csv(path + "/netflix_titles_cleaned.csv")
df.head()
```

```python
df.info()
df.columns
```

```python
df['type'].value_counts()
```

```python
df['listed_in'].value_counts().head(10)
```

```python
df['release_year'].value_counts().sort_index()
```

```python
df['country'].value_counts().head(10)
```

```python
import matplotlib.pyplot as plt

# Movies vs TV Shows
df['type'].value_counts().plot(kind='bar')
plt.title("Movies vs TV Shows")
plt.show()
```

```python
df['release_year'].value_counts().sort_index().plot()
plt.title("Content Over Years")
plt.show()
```

```python
df['country'].value_counts().head(10).plot(kind='bar')
plt.title("Top 10 Content Producing Countries")
plt.show()
```

```python
df['listed_in'].value_counts().head(10).plot(kind='bar')
plt.title("Top Genres on Netflix")
plt.show()
```

```python
df['type'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title("Movies vs TV Shows")
plt.show()
```

# Key Insights:

# 1. Netflix has significantly more movies than TV shows, indicating a stronger focus on film content.

# 2. Content production increased rapidly after 2015, showing Netflix's global expansion strategy.

# 3. The United States dominates content production, followed by other countries.

# 4. Certain genres like Drama and International Movies are most common.
